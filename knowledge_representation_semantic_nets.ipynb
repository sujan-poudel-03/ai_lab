```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes and edges
G.add_edge("Nepal", "Kathmandu", label="capital")
G.add_edge("India", "New Delhi", label="capital")
G.add_edge("China", "Beijing", label="capital")
G.add_edge("Nepal", "India", label="neighbor")
G.add_edge("Nepal", "China", label="neighbor")

```

```python
G.edges()
```

```python
G.nodes()
```

```python
# Draw the graph
plt.figure(figsize=(10, 6))  # Adjust the size here
pos = nx.spring_layout(G, seed=42)  # Set seed for reproducibility
nx.draw(G, pos, with_labels=True, node_size=1500, node_color="lightblue", font_size=12, font_weight="bold", arrowsize=20)

# Add edge labels
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title("Knowledge Representation: Countries and Capitals")
plt.show()
```

# Frame representation

```python
class CarFrame:
    def __init__(self, make, model, year, color, price):
        self.make = make
        self.model = model
        self.year = year
        self.color = color
        self.price = price

# Create instances of cars
car1 = CarFrame(make="Toyota", model="Corolla", year=2018, color="Red", price=15000)
car2 = CarFrame(make="Honda", model="Civic", year=2019, color="Blue", price=18000)
car3 = CarFrame(make="Ford", model="Fusion", year=2017, color="Black", price=16000)

# Accessing attributes of a car
print("Car 1 - Make:", car1.make)
print("Car 2 - Model:", car2.model)
print("Car 3 - Year:", car3.year)
```

# Conceptual Dependencies

```python
class Concept:
    def __init__(self, name):
        self.name = name

class Relationship:
    def __init__(self, name, source, target):
        self.name = name
        self.source = source
        self.target = target

class CDRepresentation:
    def __init__(self):
        self.concepts = []
        self.relationships = []

    def add_concept(self, name):
        self.concepts.append(Concept(name))

    def add_relationship(self, name, source, target):
        self.relationships.append(Relationship(name, source, target))

    def __str__(self):
        concept_names = {concept.name: concept.name for concept in self.concepts}
        return f"Concepts: {list(concept_names.keys())}\nRelationships: {[(rel.name, rel.source, rel.target) for rel in self.relationships]}"

# Example usage:
cd_rep = CDRepresentation()
cd_rep.add_concept("John")
cd_rep.add_concept("book")
cd_rep.add_concept("Mary")
cd_rep.add_relationship("give", "John", "book")
cd_rep.add_relationship("to", "book", "Mary")

print(cd_rep)
```
